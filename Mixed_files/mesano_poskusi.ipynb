```python
import time
import traceback

def print_lowest_queue_position_operations(credentials, order_data, sorted_operations):
    """
    This function prints the next possible operation for each module that has the lowest queuePosition
    and is not finished, but only if all previous operations for the same product are finished.
    If any operation in a module is "In progress", the module is skipped.
    """
    # Initialize a dictionary to hold operations for each module
    operations_by_module = {module: [] for module in credentials["credentials"]["module_details"]}

    # Populate operations_by_module
    for op_id, op_data in order_data["assembly"].items():
        machine_id = op_data["data"].get("machineID")
        if machine_id in operations_by_module:
            operations_by_module[machine_id].append(op_data)

    try:
        # Iterate through the modules
        for module, operations in operations_by_module.items():
            # Skip the module if any operation is "In progress"
            if any(op["metrics"].get("status") == "In progress" for op in operations):
                print(f"Skipping module {module} because it has an operation 'In progress'")
                continue

            # Sort operations by queuePosition
            sorted_module_operations = sorted(
                operations,
                key=lambda x: x["data"].get("queuePosition", float('inf'))
            )

            # Iterate through the sorted operations
            for op_data in sorted_module_operations:
                current_queue_position = op_data["data"].get("queuePosition")
                current_assembly_parent = op_data["data"].get("assemblyParent")
                current_status = op_data["metrics"].get("status")

                # Skip if the current operation is finished
                if current_status == "Finished":
                    continue

                # Check if all previous operations for the same product are finished
                all_previous_finished = True
                for group in sorted_operations:
                    for op in group:
                        if (op["data"].get("assemblyParent") == current_assembly_parent and
                            op["data"].get("queuePosition", float('inf')) < current_queue_position):
                            if op["metrics"].get("status") != "Finished":
                                all_previous_finished = False
                                print(f"Debug - Previous operation not finished: {op['data'].get('uniqueOpID')}, Status: {op['metrics'].get('status')}")
                                break  # Break if any previous operation is not finished

                # Print if the current operation is not finished and all previous operations are finished
                if all_previous_finished:
                    print(f"Module: {module}")
                    print(f"Operation ID: {op_data['data'].get('uniqueOpID')}")
                    print(f"Queue Position: {current_queue_position}")
                    print(f"Status: {current_status}")
                    print("---")
                    break  # Move to the next module after printing the first valid operation
                else:
                    print(f"Debug - Operation {op_data['data'].get('uniqueOpID')} not printed. All previous finished: {all_previous_finished}")

    except Exception as e:
        print(f"Exception: {e}")
        traceback.print_exc()



```
